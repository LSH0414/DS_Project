```python
cate = {
    0 : [],
    1 : [],
    2 : [],
    3 : []
}
for name in info_df['상품분류명']:
    for idx, c in enumerate(name.split('>')):
        cate[idx].append(c)

```

```python
tmp = []
for txt in info_df['국가정보'].unique():
    tmp.append(divNationInfo(txt))
```

```python
def divNationInfo(txt):
    # result = re.sub('\(국산\)', '', txt)

    to_create, to_make = '', ''

    if '원산지' in txt:
        to_create = re.search('(?<=원산지=)[가-힣ㄱ-ㅎ]+', txt).group(0)
    if '제조국' in txt:
        to_make = re.search('(?<=제조국=)[가-힣ㄱ-ㅎ]+', txt).group(0)
    
    return to_create, to_make
```

```python
create_lst, make_lst = [], []
for idx, row in info_df.iterrows():
    tmp_1, tmp_2 = divNationInfo(row['국가정보'])
    create_lst.append(tmp_1)
    make_lst.append(tmp_2)
```

```python
info_df['국가정보_원산지'], info_df['국가정보_제조국'] = create_lst, make_lst
```

```python
def divCompanyName(txt):
    name = re.search('(?<=이름=)[가-힣 a-zA-Z()]+', txt)
    if name:
        result = name.group(0)
    else:
        result = ''
    return result
```

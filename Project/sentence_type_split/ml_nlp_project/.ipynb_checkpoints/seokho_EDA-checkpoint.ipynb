형태소 추출기에서 명사로만 추출할 경우 극성 표현이 대부분이 생략되기 때문에 속도가 느리더라도 행태소 추출을 진행.  
형태소 추출을 진행하기 때문에 약 2만개의 문장을 분리할 것이기 때문에 속도 측면에서 가장 우수한 mecab을 사용하고 싶었으나 단어를 너무 잘게 분리 시킨다는 점에서 사용할 수 없었고 가능한 좋은 분리가 되고 있던 Okt를 사용하기로 함.

```python
def text_cleaning(text):
    # 불용어 사전 불러오기
    stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()

    custom_stopwords = ['의','가','이','은','들','는','좀','과','도','를', '을', '으로', '자', '에', '와',
            '것', '등', '년', '월', '일', '수', '그', '중','명','때', '및', '했다','에서', '로', '이다', '까지', '에는']
            
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 정규 표현식 처리 한글이 아닌 글자는 제거
    result = hangul.sub('', str(text))
    # mecab = Mecab()
    okt = Okt()
    morphs = okt.morphs(result) # 형태소 추출
    morphs = [x for x in nouns if x not in custom_stopwords]  # 커스텀 불용어 제거
    morphs = [x for x in nouns if x not in stopwords]  # 불용어 제거
    return morphs
```

### 극성 구분
 - 방법1. 긍정, 부정, 미정의 단어 사전을 각각 만들고 중복되는 단어를 제거한다.

```python
pol_list = [train_pos, train_nega, train_neu]
```

```python
len(train_pos), len(train_nega), len(train_neu)
```

```python
# 극성 별 단어 빈도수 만들기

pos_dict = counts_pos
nega_dict = counts_nega
neu_dict = counts_neu
```

```python
len(pos_dict), len(nega_dict), len(neu_dict)
```

```python
# 구분된 사전에서 다른 사전과 겹치는 단어는 제거
pop_list = []
for key in pos_dict.keys():
    isPop = False
    if nega_dict[key] != 0 :
        isPop = True
        nega_dict.pop(key)
    
    if neu_dict[key] != 0:
        isPop = True
        neu_dict.pop(key)

    if isPop:
        pop_list.append(key)

for p in pop_list:
    pos_dict.pop(p)

pop_list.clear()

for key in nega_dict.keys():
    if neu_dict[key] != 0:
        neu_dict.pop(key)
        pop_list.append(key)

for p in pop_list:
    nega_dict.pop(p)
    
```

```python
for pol_df in pol_list:
    corpus = "".join(pol_df['문장'].tolist())
    if pol_df.iloc[0]['극성'] == '긍정':
        pos_dict = Counter(text_cleaning(corpus))
    elif pol_df.iloc[0]['극성'] == '부정':
        nega_dict = Counter(text_cleaning(corpus))
    else:
        neu_dict =  Counter(text_cleaning(corpus))
```

```python
# 구분된 사전에서 다른 사전과 겹치는 단어는 제거

pop_list = []
for key, value in pos_dict.items():
    isPop = False
    if nega_dict[key] != 0 :
        if value < nega_dict[key]:
            isPop = True
        else:
            nega_dict.pop(key)
    
    if neu_dict[key] != 0:
        if value < neu_dict[key]:
            isPop = True
        else:
            neu_dict.pop(key)

    if isPop:
        pop_list.append(key)

for p in pop_list:
    pos_dict.pop(p)

pop_list.clear()

for key, value in nega_dict.items():
    if neu_dict[key] != 0:
        if value < neu_dict[key]:
            neu_dict.pop(key)
        else:
            pop_list.append(key)

for p in pop_list:
    nega_dict.pop(p)
    
```

#### 횟수를 통한 분류

```python
# 다시 극성 별 사전 정의
for pol_df in pol_list:
    corpus = "".join(pol_df['문장'].tolist())
    if pol_df.iloc[0]['극성'] == '긍정':
        pos_dict = (text_cleaning(corpus))
    elif pol_df.iloc[0]['극성'] == '부정':
        nega_dict = (text_cleaning(corpus))
    else:
        neu_dict = (text_cleaning(corpus))
```

```python
# 주어진 문장의 단어들을 극성별로 사전에 등록되어 있는 갯수 카운팅
def getCnt(df, pos_dict, nega_dict, neu_dict=[]):
    pos_cnt = []
    nega_cnt = []
    neu_cnt = []
    for idx, row in df.iterrows():
        corpus = "".join(row['문장'])
        nouns = text_cleaning(corpus)

        cnt1 = 0
        cnt2 = 0
        cnt3 = 0
        
        for word in nouns:
            if word in pos_dict:
                cnt1+=1
            if word in nega_dict:
                cnt2+=1
            if word in neu_dict:
                cnt3+=1

        pos_cnt.append(cnt1), nega_cnt.append(cnt2), neu_cnt.append(cnt3)

    cnt_df = pd.DataFrame({
        'pos' : pos_cnt,
        'nega' : nega_cnt,
        'neu' : neu_cnt
    })

    return cnt_df

train_cnt = getCnt(train, pos_dict, nega_dict, neu_dict)
```

#### 사전별 단어 중요도 정리

```python
pos_corpus = "".join(train_pos['문장'].tolist())
nega_corpus = "".join(train_nega['문장'].tolist())
neu_corpus = "".join(train_neu['문장'].tolist())

pos_corpus = " ".join(text_cleaning(pos_corpus))
nega_corpus = " ".join(text_cleaning(nega_corpus))
neu_corpus = " ".join(text_cleaning(neu_corpus))


tfdif = TfidfVectorizer(token_pattern=r'\w{1,}') # tfidf는 기본적으로 한 글자는 제외하고 중요도를 설정하기 때문에 한 글자도 중요도 설정해 주는 옵션
pos_tfidf = tfdif.fit_transform([pos_corpus])
pos_voca = tfdif.vocabulary_
nega_tfidf = tfdif.fit_transform([nega_corpus])
nega_voca = tfdif.vocabulary_
neu_tfidf = tfdif.fit_transform([neu_corpus])
neu_voca = tfdif.vocabulary_


pos_prob = {x : y for x, y in zip(pos_tfidf.indices, pos_tfidf.data)}
nega_prob = {x : y for x, y in zip(nega_tfidf.indices, nega_tfidf.data)}
neu_prob = {x : y for x, y in zip(neu_tfidf.indices, neu_tfidf.data)}
```

```python
# 문장의 극성별 단어 중요도 합 DF
def getProb(df, pos_prob, nega_prob, neu_prob = {}):
    pp = []
    np = []
    neup = []
    cnt = 0
    for idx, row in df.iterrows():
        corpus = "".join(row['문장'])
        nouns = text_cleaning(corpus)

        pp_sum = 0
        np_sum = 0
        neup_sum = 0

        
        for word in nouns:
            try:
                if word in pos_dict:
                    pp_sum += pos_prob[pos_voca[word]]
                if word in nega_dict:
                    np_sum += nega_prob[nega_voca[word]]

                if len(neu_prob) > 0:
                    if word in neu_dict:
                        neup_sum += neu_prob[neu_voca[word]]
            except:
                cnt +=1
                continue
            
        pp.append(pp_sum), np.append(np_sum), neup.append(neup_sum)

    cnt_df = pd.DataFrame({
        'pos' : pp,
        'nega' : np,
        'neu' : neup
    })

    print(cnt)

    return cnt_df

train_prob = getProb(train, pos_prob, nega_prob, neu_prob)
```

#### 모델 예측
- 통계적 분류(횟수)

```python
# 타겟 데이터 라벨 인코딩

from sklearn.preprocessing import LabelEncoder



e = LabelEncoder()
e.fit(train['극성'])
```

```python
train_cnt['극성'] = train['극성']
```

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, f1_score
```

```python
X = train_cnt[train_cnt.columns[:3]]
y = e.transform(train_cnt['True'])



X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13, test_size=0.2)


tree = DecisionTreeClassifier()
rf = RandomForestClassifier()
knn = KNeighborsClassifier(n_neighbors=10)

tree.fit(X_train, y_train)
rf.fit(X_train, y_train)
knn.fit(X_train, y_train)

pred_tree = tree.predict(X_test)
pred_rf = rf.predict(X_test)
pred_knn = knn.predict(X_test)

print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))
```

```python
#Sampling
from imblearn.over_sampling import SMOTE, RandomOverSampler # Over
from imblearn.under_sampling import RandomUnderSampler # Under
```

```python
smote = SMOTE()
ros = RandomOverSampler()
rus = RandomUnderSampler()


sm_X_train, sm_y_train = smote.fit_resample(X_train, y_train)
ros_X_train, ros_y_train = ros.fit_resample(X_train, y_train)
rus_X_train, rus_y_train = rus.fit_resample(X_train, y_train)
```

```python
# 모델 재평가 - Over sampling - Smote
tree.fit(sm_X_train, sm_y_train)
rf.fit(sm_X_train, sm_y_train)
knn.fit(sm_X_train, sm_y_train)

pred_tree = tree.predict(X_test)
pred_rf = rf.predict(X_test)
pred_knn = knn.predict(X_test)

print(" OVER SAMPLING [ SMOTE ]")
print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))
```

```python
# 모델 재평가 - Over sampling - RandomOVERSAMPLEING
tree.fit(ros_X_train, ros_y_train)
rf.fit(ros_X_train, ros_y_train)
knn.fit(ros_X_train, ros_y_train)

pred_tree = tree.predict(X_test)
pred_rf = rf.predict(X_test)
pred_knn = knn.predict(X_test)

print(" OVER SAMPLING [ RANDOM OVERSAMPLING ]")
print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))
```

```python
# 모델 재평가 - Under sampling - RANDOMUNDERSAMPLING
tree.fit(rus_X_train, rus_y_train)
rf.fit(rus_X_train, rus_y_train)
knn.fit(rus_X_train, rus_y_train)

pred_tree = tree.predict(X_test)
pred_rf = rf.predict(X_test)
pred_knn = knn.predict(X_test)

print(" Under SAMPLING [ RANDOM UNDER SAMPLING ]")
print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))
```

```python
def getPredict(df, target=train['극성']):
    X = df
    y = e.transform(target)


    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13, test_size=0.2)


    tree = DecisionTreeClassifier()
    rf = RandomForestClassifier()
    knn = KNeighborsClassifier(n_neighbors=10)

    tree.fit(X_train, y_train)
    rf.fit(X_train, y_train)
    knn.fit(X_train, y_train)

    pred_tree = tree.predict(X_test)
    pred_rf = rf.predict(X_test)
    pred_knn = knn.predict(X_test)

    print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
    print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
    print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))
```

```python
def getModel(df):
    X = df.drop('극성', axis=1)
    y = e.transform(df['극성'])

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13, test_size=0.2)

    over_sampling_instance = SMOTE()

    o_X_train, o_y_train = over_sampling_instance.fit_resample(X_train, y_train)


    tree = DecisionTreeClassifier()
    rf = RandomForestClassifier()
    knn = KNeighborsClassifier(n_neighbors=10)

    tree.fit(X_train, y_train)
    rf.fit(X_train, y_train)
    knn.fit(X_train, y_train)


    # tree.fit(o_X_train, o_y_train)
    # rf.fit(o_X_train, o_y_train)
    # knn.fit(o_X_train, o_y_train)

    pred_tree = tree.predict(X_test)
    pred_rf = rf.predict(X_test)
    pred_knn = knn.predict(X_test)

    print('Decision accuracy_score : ', accuracy_score(y_test, pred_tree))
    print('RandomForest accuracy_score : ', accuracy_score(y_test, pred_rf))
    print('KNN accuracy_score : ', accuracy_score(y_test, pred_knn))

    return tree, rf, knn
```

```python
df = pd.read_csv('events_cnt_div.csv', index_col=0)
event_df = pd.read_csv('vocie_fishing.csv', index_col=0)
```

```python
df = df[['year','month','address','addr1', 'addr2', 'name', 'lat', 'lng', 
       'events_in_100m', 'events_in_500m', 'events_in_1km',]]
```

```python
df
```

```python
# 충남, 대전, 세종특별자치시
df_CN = df[df['addr1']=='충남']
```

```python
df_CN[df_CN['events_in_100m']>=1]
```

```python
CN_addr2 = df_CN.groupby('addr2')['events_in_100m', 'events_in_500m', 'events_in_1km'].sum()
```

```python
CN_addr2
```

```python
top_region = list(set(CN_addr2.sort_values('events_in_100m', ascending=False).index[:5].tolist() + CN_addr2.sort_values('events_in_500m', ascending=False).index[:5].tolist() + CN_addr2.sort_values('events_in_1km', ascending=False).index[:5].tolist()))
```

```python

CN_top_region = CN_addr2.loc[top_region]

fig = px.bar(data_frame=CN_top_region, barmode='group')

fig.show()
```

```python
event_df.head()
```

```python
cnt_event = event_df.groupby(['접수완료일']).size().reset_index().rename(columns={0:'cnt'})
```

```python
cnt_event = cnt_event.set_index('접수완료일')
```

```python
roll_mean30 = pd.Series.rolling(cnt_event['cnt'], window=30, center = False).mean()
roll_mean120 = pd.Series.rolling(cnt_event['cnt'], window=120, center = False).mean()
```

```python
cnt_event
```

```python

import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig = plt.figure(figsize = (20, 10),dpi=1000)
chart = fig.add_subplot(1,1,1)

chart.plot(roll_mean30, color='red' , label='1달 이동평균')
chart.plot(roll_mean120, color='orange' , label='4달 이동평균')

plt.xticks([i for i in range(0, len(cnt_event), 120)])
plt.legend(loc = 'best')

plt.savefig('추세선.png',transparent = True)
```

```python
to_concat = to_concat.reset_index()
```

```python
to_concat = to_concat[(to_concat['addr1'] == '충남') | (to_concat['addr1'] == '대전') | (to_concat['addr1'] == '세종특별자치시')]
```

```python
pop = pop.drop(index=35).reset_index()
```

```python
to_concat = to_concat.rename(columns= {'addr22':'주소2'})
```

```python
pop.columns, to_concat.columns
```

```python
concats = pop.merge(to_concat, how='inner', on='주소2')
```

```python
concats.head()
```

```python
concats.columns
```

```python
concats = concats.drop(['index', 'addr1'], axis=1)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

target = concats[concats['주소']=='대전광역시'].sort_values('cnt', ascending=False)

fig.add_trace(go.Bar(x=target['주소2'].values, y=target['cnt'].values, text=target['cnt'].values, textposition="outside", name="신고 수"))

fig.update_layout(
    yaxis1=dict(
        title="신고 수",
        titlefont=dict(
            # color="#ff7f0e"
        ),
        tickfont=dict(
            # color="#ff7f0e"
        ),
        # anchor="free",
        # overlaying="y",
        side="left",
        # position=0.15
    ),
    yaxis2=dict(
        title="인구수",
        titlefont=dict(
            # color="#d62728"
        ),
        tickfont=dict(
            # color="#d62728"
        ),
        anchor="x",
        overlaying="y",
        side="right"
    )
    )
# fig.update_yaxes(range=[0,13000])

fig.add_trace(go.Line(x=target['주소2'].values, y=target['총인구수'].values, name="인구수"),secondary_y=True)

fig.show()  
```

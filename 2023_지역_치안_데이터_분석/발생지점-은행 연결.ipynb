```python
def get_swap_dict(d):
    return {v: k for k, v in d.items()}
```

```python
change_name_dict = {
    '대전광역시' : '대전',
    '충청남도' : '충남',
    '경기도' : '경기',
    '강원도' : '강원',
    '전라남도': '전남',
    '충청북도' : '충북',
    '서울특별시' : '서울',
    '광주광역시' : '광주',
    '경상북도' : '경북',
    '부산광역시' : '부산',
    '경상남도' : '경남',
    '전라북도' : '전북',
    '인천광역시' : '인천',
    '대구광역시' : '대구',
    '울산광역시' : '울산',
    '세종특별자치시' : '세종특별자치시'
    }
```

```python
from numpy import sin, cos, arccos, pi, round

def rad2deg(radians):
    degrees = radians * 180 / pi
    return degrees

def deg2rad(degrees):
    radians = degrees * pi / 180
    return radians

def getDistanceBetweenPointsNew(latitude1, longitude1, latitude2, longitude2, unit = 'kilometers'):
    
    theta = longitude1 - longitude2
    
    distance = 60 * 1.1515 * rad2deg(
        arccos(
            (sin(deg2rad(latitude1)) * sin(deg2rad(latitude2))) + 
            (cos(deg2rad(latitude1)) * cos(deg2rad(latitude2)) * cos(deg2rad(theta)))
        )
    )
    
    if unit == 'miles':
        return round(distance, 2)
    if unit == 'kilometers':
        return round(distance * 1.609344, 2)
```

```python
region_dict = get_swap_dict(change_name_dict)
```

```python
df['year'] = df['접수완료일'].apply(lambda x : x[:4])
```

```python
df['year'] = df['year'].astype('int')
```

```python
bank['year'] = 0
bank['month'] = 0
```

```python
bank['events_in_100m'] = 0
bank['events_in_300m'] = 0
bank['events_in_500m'] = 0
bank['events_in_1km'] = 0

event_cnt_df = pd.DataFrame(columns=bank.columns)
```

```python
df['month'] = df['접수완료일'].apply(lambda x : x[5:7])
```

```python
from tqdm import tqdm 
```

```python
for idx in range(len(bank)):
    if type(bank['addr1'][idx]) != str : continue

    addr1 = region_dict[str(bank['addr1'][idx])]
    target = df[df['발생지점'].str.contains(addr1)]
    target = target[target['발생지점'].str.contains(str(bank['addr2'][idx]))]
    
    

    if len(target):
        years = target['year'].unique()

        for i in years:
            cnt_100m = 0
            cnt_500m = 0
            cnt_1km = 0
            
            add_df = bank.iloc[idx]
            add_df['year'] = i

            target_year = target[target['year']==i]
            lats = target_year['발생좌표Y'].values
            lngs = target_year['발생좌표X'].values
            for lat,lng in zip(lats, lngs):

                distance = getDistanceBetweenPointsNew(lat,lng, bank['lat'][idx], bank['lng'][idx])
                
                if distance <= 0.1:
                    cnt_100m +=1
                elif distance <= 0.5:
                    cnt_100m += 1
                    cnt_500m += 1
                elif distance <= 1:
                    cnt_100m += 1
                    cnt_500m += 1
                    cnt_1km +=1
                
            
            add_df['events_in_100m'] = cnt_100m
            add_df['events_in_500m'] = cnt_500m
            add_df['events_in_1km'] = cnt_1km

            event_cnt_df = event_cnt_df.append(add_df, ignore_index=True)
        # break
            
    
```

```python
for idx in tqdm(range(len(bank))):
    if type(bank['addr1'][idx]) != str : continue

    addr1 = region_dict[str(bank['addr1'][idx])]
    target = df[df['발생지점'].str.contains(addr1)]
    target = target[target['발생지점'].str.contains(str(bank['addr2'][idx]))]
    
    

    if len(target):
        years = target['year'].unique()

        for i in years:
            add_df = bank.iloc[idx]
            add_df['year'] = i

            target_year = target[target['year']==i]
            months = target_year['month'].unique()
            for j in months:
                cnt_100m = 0
                cnt_300m = 0
                cnt_500m = 0
                cnt_1km = 0
                
                target_month = target_year[target_year['month']== j]
                lats = target_month['발생좌표Y'].values
                lngs = target_month['발생좌표X'].values
                add_df['month'] = j
                for lat,lng in zip(lats, lngs):

                    distance = getDistanceBetweenPointsNew(lat,lng, bank['lat'][idx], bank['lng'][idx])
                    
                    if distance <= 0.1:
                        cnt_100m +=1
                        cnt_300m +=1
                        cnt_500m +=1
                        cnt_1km +=1
                    elif distance <= 0.3:
                        cnt_300m +=1
                        cnt_500m += 1
                        cnt_1km +=1
                    elif distance <= 0.5:
                        cnt_500m += 1
                        cnt_1km +=1
                    elif distance <= 1:
                        cnt_1km +=1
                    
                
                add_df['events_in_100m'] = cnt_100m
                add_df['events_in_300m'] = cnt_300m
                add_df['events_in_500m'] = cnt_500m
                add_df['events_in_1km'] = cnt_1km

                event_cnt_df = event_cnt_df.append(add_df, ignore_index=True)
        # break
```

```python
event_cnt_df.to_csv('events_cnt_div.csv')
```
